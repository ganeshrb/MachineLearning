### LOOPs in SAS:

#### Find the sum of first 15 natural numbers

```sas
* do while loop;
DATA SampleLoop;

SUM = 0;
VAR = 0; 

Do While(VAR < 16);
SUM = SUM + VAR;
VAR = VAR + 1;
END;

PROC PRINT DATA = SampleLoop;
Run;
```

```sas
* Do index loop;

DATA SampleLoop;
SUM = 0;
VAR = 0;

Do VAR = 1 to 15;
SUM = SUM + VAR;
END;

PROC PRINT DATA = SampleLoop;
Run;
```

```sas
* Do until loop;
DATA SampleLoop;
SUM=0;
VAR=0;

Do Until(VAR > 15);
SUM = SUM + VAR;
VAR = VAR + 1;
END;

PROC PRINT DATA = SampleLoop;;
Run;
```

#### SAS Histogram with Normal Curve

```sas
TITLE 'Summary of Weight Variable (in pounds)';
PROC UNIVARIATE DATA = sashelp.class NOPRINT;
HISTOGRAM weight / NORMAL; 
RUN;
```

```sas
TITLE 'Summary of Height Variable';
PROC UNIVARIATE DATA = sashelp.class NOPRINT;
var Height;
HISTOGRAM / NORMAL; 
RUN;
```
